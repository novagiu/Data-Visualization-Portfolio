```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

degree_gender_ratio = pd.read_csv('/percent-degrees-conferred-women-usa.csv')

degree_gender_ratio.head()
majors=degree_gender_ratio.columns[1:]
x=degree_gender_ratio['Year']
for major in majors:
    y=degree_gender_ratio[major]
    plt.plot(x,y,label=major)
plt.ylim(0,40)
plt.legend(bbox_to_anchor=(1.40, 1),
           title= "Majors",
           loc='upper right',
           prop={'size':6})
plt.xlabel('Year')
plt.ylabel('Female Ratio')
plt.title('Gender Ratio of Different Majors')
plt.show()
```
